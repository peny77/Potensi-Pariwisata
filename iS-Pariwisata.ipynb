## Jenis Obyek Wisata

### `Pengelompokkan Jenis Obyek Wisata`

> Pengelompokkan ini bertujuan agar kita mengetahui ada berapa sebenarnya jenis obyek wisata yang ada di **`Kabupaten Klaten`** ketika sudah di kelompokkan berdasarkan jenisnya.

```python
ab = df.groupby(['Jenis Obyek Wisata', 'Nama Obyek Wisata','Lokasi']).count()
ab
```

> Dari catatan `dataset` atau data awal yang mungkin terasa tidak beraturan dan tidak semua orang dapat membaca dan memahami, namun setelah di kelompokkan menurut jenisnya akan terlihat mudah dibaca dan dipahami.

### `Visualisasi Jenis Obyek Wisata`

> Dari pengelompokkan yang telah dilakukkan sebelumnya untuk lebih jelas dan paham lagi kita akan gambarkan kedalam sebuah visualisasi `bar chart` di bawah ini

```python
import matplotlib.pyplot as plt

df['Jenis Obyek Wisata'].value_counts().head(30).sort_values(ascending=True).plot(kind='barh',
                                                      figsize=(11,7),
                                                      title='Jenis Obyek Wisata',
                                                      color='purple');
```

> Hasil dari gambaran visualisasi di atas dapat kita ketahui bersama bahwa **`Kabupaten Klaten`** setidaknya mempunyai 13 jenis obyek wisata yang potensial di antaranya adalah
1. Potensi Wisata Alam berjumlah 6
2. Lima Wisata Sejarah dan Religi
3. Empat Event Tradisional
4. Dua Wisata Air
5. dan Wisata Budaya dan Industri, Wisata Budaya Kerajinan dan Kerajinan, Wisata Air dan Perikanan Nila, Pemancingan, Industri Lurik, Wisata Budaya dan Kerajinan Batik, Wisata Religi dan Alam, Wisata Air dan Snarkling, Wisata Budaya dan Outbond masing-masing satu.

## TOP 3

```python
top_tiga = df['Jenis Obyek Wisata'].value_counts()
satu = top_tiga[0]
dua = top_tiga[1]
tiga = top_tiga[2]
```

```python
import numpy as np

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ['Wisata Alam',
          'Wisata Sejarah dan Religi',
          'Event Tradisional']

data = [satu, dua, tiga]

wedges, texts = ax.pie(data, wedgeprops=dict(width=0.5), startangle=-20)

bbox_props = dict(boxstyle="square,pad=0.5", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("TOP 3 Jenis Obyek Wisata")

plt.show()
```

> Visualisasi `TOP 3` yang dimaksudkan untuk mengetahui tiga besar jenis obyek wisata yang di miliki oleh **`Kabupaten Klaten`** untuk dikembangkan bahkan mungkin untuk ditambah. Wisata Alam kemungkinan masih bisa dibuat atau ditambah dan Event Tradisional juga memungkinkan untuk bisa digali lebih dalam lagi dengan tidak menutup kemungkinan obyek wisata yang lain untuk bertambah dan naik ke TOP 3 serta menggeser posisi tiga besar saat ini.
